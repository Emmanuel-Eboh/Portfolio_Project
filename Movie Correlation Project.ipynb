```python
# Import Libraries

import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None


# Read in the data

df = pd.read_csv(r"C:\Users\EMMADEX\Documents\Portfolio Project\movies.csv")
```

```python
# Lets Look at the data

df.head()
```

```python
# We need to see if we have any missing data
# Let's loop through the data and see if there is anything missing

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
# Looking at the data types for our columns

df.dtypes
```

```python
# Fixing IntcastingNaNError

df = df.fillna(0)
```

```python
# Change data types of budget and gross columns

df['budget'] = df['budget'].astype('int64')

df['gross'] = df['gross'].astype('int64')
```

```python
df
```

```python
# Converting released column to string & creating new column (yearcorrect)

df['yearcorrect'] = df['released'].astype(str).str[:4]

df
```

```python
# Ordering our data

df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
# Drop any duplicates

df.drop_duplicates()
```

```python
# Scatter plot with budget vs gross

plt.scatter(x=df['budget'], y=df['gross'])

plt.title('Budget vs Gross Earnings ')

plt.xlabel('Gross Earnings')

plt.ylabel('Budget for Film')

plt.show()

```

```python
# Plot Budget vs Gross using seaborn

sns.regplot(x="budget", y="gross", data=df, scatter_kws={'color': 'red'}, line_kws={'color': 'blue'})
```
